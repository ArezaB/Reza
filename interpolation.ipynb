Here, I put 3 versions of an interpolation code:
1. It accepts a list to interpolate
2. It accepts a list via input function to interpolate
3. It reads a list from an excel file and does the interpolation

```python
import pandas as pd
mydata=pd.read_excel('interp.xlsx')
```

```python
#by this code you interpolate a list (x_hat)
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = mydata.Time
y = mydata.Flow

f = interp1d(x, y)
x_hat=[0.6,1,2,6,11.8,43,51.5]
#x_hat=input()
#x_hat = []
#n = int(input("Enter the list size "))

#print("\n")
#for i in range(0, n):
    #print("Enter number at index", i, )
    #item = float(input())
    #x_hat.append(item)
#print("User list is ", number_list)

for i in f(x):
 y_hat = f(x_hat)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(x_hat, y_hat, 'ro')
plt.title('Linear Interpolation at x_hat')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#by this code you interpolate a list of values using input function
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = mydata.Time
y = mydata.Flow

f = interp1d(x, y)
#x_hat=[0.6,1,2,6,11.8,43,51.5]
#x_hat=input()
x_hat = []
n = int(input("Enter the list size "))

print("\n")
for i in range(0, n):
    #print("Enter number at index", i, )
    item = float(input())
    x_hat.append(item)
#print("User list is ", x_hat)

for i in f(x):
 y_hat = f(x_hat)
print(y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(x_hat, y_hat, 'ro')
plt.title('Linear Interpolation at x_hat')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
