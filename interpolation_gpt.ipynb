```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load the data
data_file = '/content/drive/My Drive/testdata.xlsx'
df = pd.read_excel(data_file)
x = df.iloc[:, 0].values
y = df.iloc[:, 1].values

# Load the interpolation points
interp_file = '/content/drive/My Drive/interp.xlsx'
df_interp = pd.read_excel(interp_file)
x_interp = df_interp.iloc[:, 0].values

# Interpolate the data
f = interp1d(x, y, kind='linear')
y_interp = f(x_interp)

# Plot the data and interpolated points
fig, ax = plt.subplots()
ax.plot(x, y, 'o-', label='Data')
ax.plot(x_interp, y_interp, 'ro', label='Interpolated Points')

# Label the plot
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Linear Interpolation')
ax.legend()

# Print the interpolated points in a table and save to file
df_interp['Interpolated y'] = y_interp
df_interp.to_excel(interp_file, index=False) # add this line to save the DataFrame to file

# Display the plot
plt.show()
```
